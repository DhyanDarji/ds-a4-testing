```python
import matplotlib.pyplot as plt 

```

```python
nik_results = [0.0070037841796875, 0.015417759418487548, 0.0456234073638916, 0.16984165363311768, 0, 0]
hugh_results = [0.007946491241455078, 0.014754884243011475, 0.051265509843826294, 0.15101676235198974, 0.5449067010188102, 1.647655991911888]
dhyan_results = [0.03618907928466797, 0.018139834403991698, 0.055646658420562746, 0.17166674489974976, 0.6014623495173455, 1.6682640291621684]
x_axis = [1, 100, 1000, 10000, 100000, 1000000]
```

```python
plt.figure(figsize=(10,7))

hugh_x_axis = [0,1,2,3,4,5]
nik_x_axis = [x-0.2 for x in hugh_x_axis]

dhyan_x_axis = [x+0.2 for x in hugh_x_axis]


plt.bar(nik_x_axis, nik_results, color="green", label="Nik", width=0.2)
plt.bar([0,1,2,3,4,5], hugh_results, color="orange", label="Hugh", width=0.2)
plt.bar(dhyan_x_axis, dhyan_results, color="blue", label="Dhyan", width=0.2)
plt.title("Average Response Time vs Number of Requests Sent")
plt.ylabel("Average Resposne (seconds)")
plt.xlabel("Requests (1e6)")
plt.xticks([0,1,2,3,4,5], x_axis)
plt.legend()
plt.savefig("experiment2.png", dpi=1000)
plt.show()
```
